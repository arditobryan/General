```
#naive recursion
def Fibonacci(n):
  if n <= 1:
    return n
  return(Fibonacci(n-1) + Fibonacci(n-2))
Fibonacci(100)
```

Memoization

image.png

```
#dynamic programming, memoization
fib_cache = {}

def fib(n):
#if n already exist in store
	#4. MEMOIZATION RETRIEVAL
	if n in fib_cache:
		value = fib_cache[n]
		return value

#if n does not exist in store
	#5. EXIT
	if n == 1:
		return 1
	if n == 2:
		return 1
	
	#2. ALGORITHM
	value = fib(n-2)+fib(n-1)
	#Because n does not exist in store, 3. MEMOIZATION STORING
	fib_cache[n] = value
	return value

#1. CALL
print(fib(1000))
```
